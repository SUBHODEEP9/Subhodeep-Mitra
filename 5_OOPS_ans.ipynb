```python
# 1. Create a parent class Animal with a method speak() that prints a generic message. Create a child class Dog that overrides the speak() method to print "Bark!".

class Animal:
  def speak(self):
    print("Animals have sound")
class Dog(Animal):
  def speak(self):
    print("Bark")
a = Animal()
a.speak()
d = Dog()
d.speak()
```

```python
# 2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both.

from abc import ABC, abstractmethod
import math

# Abstract class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

# Derived class Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

# Derived class Rectangle
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width


# Creating objects
c = Circle(5)
r = Rectangle(4, 6)

print("Circle Area:", c.area())
print("Rectangle Area:", r.area())
```

```python
''' 3. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car and further derive a class ElectricCar
      that adds a battery attribute.'''

# Parent class
class Vehicle:
    def __init__(self, type):
        self.type = type

# Child class
class Car(Vehicle):
    def __init__(self, type, brand):
        super().__init__(type)
        self.brand = brand

# Grandchild class
class ElectricCar(Car):
    def __init__(self, type, brand, battery):
        super().__init__(type, brand)
        self.battery = battery


# Creating object
e = ElectricCar("Four Wheeler", "Tesla", "75 kWh")

print("Type:", e.type)
print("Brand:", e.brand)
print("Battery:", e.battery)
```

```python
# 4. Demonstrate polymorphism by creating a base class Bird with a method fly(). Create two derived classes Sparrow and Penguin that override the fly() method.
class Bird:
  def fly(self):
    print("Birds can fly.")
class Sparrow(Bird):
  def fly(self):
    print("Sparrow can fly.")
class Penguin(Bird):
  def fly(self):
    print("Penguin can't fly.")

birds = [Sparrow(), Penguin()]

def sky(birds):
  for i in birds:
    i.fly()
sky(birds)
```

```python
# 5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance.

class BankAccount:
  def __init__(self, balance):
    self.__balance = balance
  def deposit(self,amount):
    self.__balance += amount
    print("Amount Deposited: ", amount)

  def withdraw(self, amount):
    if amount <= self.__balance:
      self.__balance -= amount
      print("Withdrawn amount: ", amount)
    else:
      print("Insufficient balance.")

  def check_balance(self):
    print("Current balance: ", self.__balance)

account = BankAccount(2000)
account.check_balance()
account.deposit(500)
account.withdraw(250)
account.check_balance()
```

```python
# 6. Demonstrate runtime polymorphism using a method play() in a base class Instrument. Derive classes Guitar and Piano that implement their own version of play().

class Instrument:
  def play(self):
    print("Instrument is playing melodiously.")
class Guitar(Instrument):
  def play(self):
    print("Guitar is playing.")
class Piano(Instrument):
  def play(self):
    print("Piano is playing.")
instruments = [Guitar(), Piano()]

def music(instruments_list):
  for i in instruments_list:
    i.play()

# Call the function that iterates through the list
music(instruments)
```

```python
# 10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two vectors.

class Vector:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def __add__(self, other):
    return Vector(self.x + other.x, self.y + other.y)

  def __str__(self):
    return f"({self.x}, {self.y})"

v1 = Vector(2, 3)
v2 = Vector(4, 5)

v3 = v1 + v2   # Operator overloading

print(v3)
```

```python
''' 14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager
        that adds a bonus to the salary.'''
class Employee:
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate
        self.bonus = 0  # Initial bonus is zero

    def calculate_base_salary(self):
        return self.hours_worked * self.hourly_rate

    def calculate_total_salary(self):
        # Returns the base pay plus whatever bonus has been assigned
        return self.calculate_base_salary() + self.bonus

class Manager(Employee):
    def __init__(self, name, hours_worked, hourly_rate):
        super().__init__(name, hours_worked, hourly_rate)

    def add_bonus(self, employee, bonus_amount):
        """The Manager interacts with an Employee object to add a bonus."""
        employee.bonus += bonus_amount
        print(f"Manager {self.name} awarded a ${bonus_amount} bonus to {employee.name}.")

# 1. Create the staff
staff_member = Employee("Mahi", 40, 25)
dept_manager = Manager("BCCI", 40, 50)

print(f"{staff_member.name}'s Base Salary: ${staff_member.calculate_base_salary()}")

dept_manager.add_bonus(staff_member, 500)

print(f"{staff_member.name}'s Total Salary (with bonus): ${staff_member.calculate_total_salary()}")
```

```python
# 16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method.

from abc import ABC, abstractmethod

class Animal:
  @abstractmethod
  def sound(self):
    pass

class Cow(Animal):
  def sound(self):
    return "Moo"
class Sheep(Animal):
  def sound(self):
    return "Baa"

c = Cow()
s = Sheep()
print(c.sound())
print(s.sound())
```

```python
# 18. Create a class House with attributes address and price. Create a derived class Mansion that adds an attribute number_of_rooms.

class House:
  def __init__(self, address, price):
    self.address = address
    self.price = price

class Mansion(House):
  def __init__(self, address, price, number_of_rooms):
    super().__init__(address, price)
    self.number_of_rooms = number_of_rooms

  def __str__(self):
    return f"Address: {self.address}, Price: {self.price}, Rooms: {self.number_of_rooms}"

m1 = Mansion("Bally, Howrah", 1200000, 6)
m2 = Mansion("Karnatak", 3500000, 8)

print(m1)
print(m2)
```
