# Python Basics Question and **Answers**

### *THEORY QUESTIONS*

---




1. **What is Python, and why is it popular?**
   
   --> Python is a high-level, interpreted programming language known for its simplicity, readability, and wide range of libraries, making it popular for beginners and professionals.

2. **What is an interpreter in Python?**
   
   --> An interpreter executes Python code line by line, converting it into machine-readable instructions.

3. **What are pre-defined keywords in Python?**
   
   --> Keywords are reserved words with special meanings in Python, such as 'if', 'else', 'for', 'while', and 'def'.

4. **Can keywords be used as variable names?**
   
   --> No, keywords cannot be used as variable names because they are reserved for Python’s syntax.

5. **What is mutability in Python?**
   
   --> Mutability refers to whether an object’s value can be changed after it is created.

6. **Why are lists mutable, but tuples are immutable?**
   
   --> Lists are designed to allow changes, while tuples are fixed to ensure data safety and faster performance.

7. **What is the difference between “==” and “is” operators in Python?**
   
   --> "==" compares values for equality, while "is" checks whether both variables refer to the same object in memory.

8. **What are logical operators in Python?**
   
   --> Logical operators are 'and', 'or', and 'not', used to combine conditional statements.

9. **What is type casting in Python?**
   
   --> Type casting is the process of converting one data type into another.

10. **What is the difference between implicit and explicit type casting?**
    
    --> Implicit casting is done automatically by Python, while explicit casting is done manually using functions like 'int()', 'float()', or 'str()'.

11. **What is the purpose of conditional statements in Python?**
    
    --> Conditional statements allow a program to execute different blocks of code based on conditions.

12. **How does the elif statement work?**
    
    --> 'elif' checks an additional condition if the previous 'i' condition is false.

13. **What is the difference between for and while loops?**
    
    --> A 'for' loop iterates over a sequence, while a 'while' loop runs as long as a condition remains true.

14. **Describe a scenario where a while loop is more suitable than a for loop.**
    
    --> A 'while' loop is suitable when the number of iterations is unknown, such as reading input until the user enters a specific value.

```python
# 5. Create a list and tuple in Python, and demonstrate how attempting to change an element works differently for each.

# Create a list
my_list = [10, 20, 30]

# Create a tuple
my_tuple = (10, 20, 30)



# Changing an element in the list
my_list[1] = 99
print("Updated list:", my_list)

# This will not show any error.
```

```python
# Trying to change an element in the tuple
my_tuple[1] = 99   # This will cause an error
print("Updated tuple:", my_tuple)


# Changing list will not show any error. But changing tuple will show error.
# This happened because, List is mutable and Tuple is immutable in nature.
```

```python
# 6. Write a function to demonstrate the behavior of mutable and immutable arguments.

def exam(num, lst):
    num += 5        # immutable
    lst.append(5)   # mutable

x = 7
y = [1, 2]

exam(x, y)

print(x)  # unchanged
print(y)  # changed
```

```python
# 8. Write a program to demonstrate the use of logical operators.

# Take input from user
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

# Logical AND
print("a > 0 and b > 0 :", a > 0 and b > 0)

# Logical OR
print("a > 0 or b > 0 :", a > 0 or b > 0)

# Logical NOT
print("not (a > 0) :", not (a > 0))

# Logic LEFT SHIFT
print("a << b :", a << b)

# Logic RIGHT SHIFT
print("a >> b :", a >> b)
```

```python
# 14.  Write a program to reverse a string using a while loop.

text = input("Enter a string: ")

reversed_text = ""
i = len(text) - 1

while i >= 0:
    reversed_text += text[i]
    i -= 1

print("Reversed string:", reversed_text)
```

```python
# 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

num = int(input("Enter a number: "))

factorial = 1
i = num

while i > 0:
    factorial = factorial * i
    i -= 1

print("Factorial of", num, "is:", factorial)
```
