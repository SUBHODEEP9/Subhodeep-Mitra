```python
# 4. Write a Python script that reads from one file and writes its content to another file.
try:
    # Open the source file in read mode
    with open("source.txt", "r") as source_file:
        content = source_file.read()

    # Open the destination file in write mode
    with open("destination.txt", "w") as destination_file:
        destination_file.write(content)

    print("Content copied successfully.")
except FileNotFoundError:
    print("Error: The file 'source.txt' was not found. Please ensure it exists in the same directory as this notebook.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
# 17. Write a Python program to create and write a list of numbers to a file, one number per line.

# 1. Create a list of numbers
numbers = [10, 20, 30, 40, 50]

# 2. Open the file in write mode using a 'with' statement
file_name = "numbers.txt"
with open(file_name, 'w') as f:
    # 3. Iterate and write each number to the file, followed by a newline
    for number in numbers:
        f.write(str(number) + '\n')

# 4. Confirmation message
print(f"Numbers successfully written to {file_name}")

# Optional: Verify the content by reading the file
print("\nContent of the created file:")
with open(file_name, 'r') as f:
    print(f.read())
```

```python
# 18. How would you implement a basic logging setup that logs to a file with rotation after 1MB?
import logging
from logging.handlers import RotatingFileHandler

handler = RotatingFileHandler("app.log", maxBytes=1000000, backupCount=3)


formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
handler.setFormatter(formatter)

# Get the logger and set level
logger = logging.getLogger()
logger.setLevel(logging.INFO)

# Add handler to logger
logger.addHandler(handler)

# Example log messages
logger.info("This is an info message")
logger.warning("This is a warning message")
logger.error("This is an error message")
```
