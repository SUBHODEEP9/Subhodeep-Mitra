2.   Describe the time complexity of accessing elements in a dictionary.

     -->

*  ** Average Case:** O(1) (Constant Time)
       Dictionaries use a hash table.

      When you access a value using a key, Python computes the hash of the key and directly finds the corresponding value.

      This makes lookups very fast, regardless of the dictionary size

*   **Worst Case: O(n)**
    This happens rarely, when hash collisions occur (multiple keys map to the same hash location).

    Python resolves collisions internally, but in extreme cases it may need to check multiple entries.

2.   Why are dictionaries considered unordered, and how does that affect data retrieval?

      --> Dictionaries do not rely on index positions but on keys. Data retrieval is unaffected because lookup uses keys, not order.


2.   Explain the difference between a list and a dictionary in terms of data retrieval.

     -->

*   A list retrieves data using index positions, so accessing an element requires knowing its numeric position, and searching for a specific value may take more time.

     List data retrieval is generally slower for large datasets because it may require sequential searching.

*   In contrast, a dictionary retrieves data using keys, allowing direct and faster access to values.

    Dictionary retrieval is faster because it uses hashing to locate values efficiently.
